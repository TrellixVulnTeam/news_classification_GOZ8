```python
import pandas as pd
import pickle
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score
```

```python
kaggle_data = pd.read_csv('kaggle_data.csv')
```

```python
kaggle_data.head()
```

```python
training_data_new = pd.concat([kaggle_data['TITLE'], kaggle_data['CATEGORY']], axis=1)
category_list_dict = {'b':0, 't':1, 'e':2, 'm':3}
training_data_new['CATEGORY'] = training_data_new['CATEGORY'].map(category_list_dict)
training_data_new
```

```python
count_vec = CountVectorizer()
X_train_counts = count_vec.fit_transform(training_data_new['TITLE'])
print(X_train_counts)
print(count_vec.vocabulary_)
print(len(count_vec.vocabulary_))
#SAVE WORD VECTOR
pickle.dump(count_vec.vocabulary_, open("count_vec.pkl","wb"))
```

```python
#TRANSFORM WORD VECTOR TO TF IDF
tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)
print(X_train_tfidf)
#SAVE TF-IDF
pickle.dump(tfidf_transformer, open("tfidf_new.pkl","wb"))
```

```python
clf_svm = svm.LinearSVC()
X_train, X_test, y_train, y_test = train_test_split(X_train_tfidf, training_data_new['CATEGORY'], test_size=0.25, random_state=42)
#clf_svm.fit(X_train, y_train)
#pickle.dump(clf_svm, open("svm_new.pkl", "wb"))
```

```python
print(X_train)
print(X_test)
```

```python
category_list_dict_rev = {0:'b', 1:'t', 2:'e', 3:'m'}
```

```python
predicted = clf_svm.predict(X_test)
result_svm = pd.DataFrame( {'true_labels': y_test,'predicted_labels': predicted})
result_svm.to_csv('res_svm_new.csv', sep = ',')

for predicted_item, result in zip(predicted, y_test):
    print(category_list_dict_rev[predicted_item], ' - ', category_list_dict_rev[result])
```

```python
print(accuracy_score(y_test, predicted))
print(f1_score(y_test, predicted, average='macro'))
print(recall_score(y_test, predicted, average='macro'))
print(precision_score(y_test, predicted, average='macro'))
```
